# Graficador de Mediciones

```python
import matplotlib.pyplot as plt
import numpy as np 
import os
import pandas as pd
```

## Directorios y carga de archivos

```python
# Directorio de entrada de datos (mediciones)
DATA_DIR = "../data"
# Directorio donde se guardan las mediciones
MEASURE_DIR = "mic_behringher_ecm8000"
# Directorio de salida de gráficos
OUTPUT_DIR = "output"
output_path = os.path.join(OUTPUT_DIR, MEASURE_DIR)
os.makedirs(output_path, exist_ok=True)
# Crear una lista que contiene los archivos .txt de la medición
files = os.listdir(os.path.join(DATA_DIR, MEASURE_DIR))
print(files[0])
```

## Análisis de Datos

```python
# Crear un dataframe a partir de un .txt (o un .csv)
path_file = os.path.join(DATA_DIR, MEASURE_DIR, files[0])
df = pd.read_csv(path_file, delimiter="\t", header=1)
df
```

## Graficar resultados preliminares

```python
# Grafico rápido del dataframe, frecuencia vs nivel dB
x_ticks = [20,100,200,500,1000,5000,10000,20000]
x_labels = ['20','100','200','500','1k','5k','10k','20k']
y_ticks = np.arange(-12,13,3)
plt.figure(figsize=(16,9))
plt.plot(df['Frequency (Hz)'], df['Magnitude (dB)'])
plt.title('Respuesta en frecuencia del micrófono Behringer ECM8000', fontsize=20)
# Etiqueta los ejes y límites
plt.xscale('log')
plt.xlim(20, 20000)
plt.ylim(-12,12)
plt.xticks(x_ticks, x_labels)
plt.yticks(y_ticks)
plt.xlabel('Frequency (Hz)', fontsize=14)
plt.ylabel('Magnitude (dB)', fontsize=14)
plt.grid(which='both', linestyle='--', linewidth=0.5)
plt.savefig(os.path.join(output_path, "behringher_0.png"), dpi=300)
plt.show()



```
